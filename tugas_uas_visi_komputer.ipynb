```python
import cv2
from google.colab.patches import cv2_imshow  # Untuk menampilkan gambar di Colab
```

```python
from google.colab import files
uploaded = files.upload()
```

```python
# Membaca gambar menggunakan OpenCV
image_path = list(uploaded.keys())[0]
image = cv2.imread(image_path)
```

```python
# Mengonversi gambar ke dalam skala abu-abu
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
```

```python
# Melakukan thresholding dengan nilai ambang 128 (bisa disesuaikan)
_, thresholded_image = cv2.threshold(gray_image, 12, 50, cv2.THRESH_BINARY)
```

```python
# Menampilkan gambar asli dan gambar setelah thresholding
cv2_imshow(image)
cv2_imshow(thresholded_image)
```
